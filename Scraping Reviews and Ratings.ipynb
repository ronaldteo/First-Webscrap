```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time

#List of string used to modify the URL
page_number_list = list(map(str,range(10,121,10)))
```

```python
#Dataframe for storing the scraped datas
import pandas as pd

headers = ["Hotel","Review","Rating"]
df = pd.DataFrame(columns = headers)
df
```

### Scraping for Ritz Carlton

```python
for num in page_number_list:
    url = f"https://www.booking.com/reviewlist.en-gb.html?aid=304142&label=gen173nr-1FCAEoggI46AdIM1gEaMkBiAEBmAEJuAEXyAEM2AEB6AEB-AEMiAIBqAIDuAKQuIqtBsACAdICJGJlNTkyNDdhLTk0NWYtNGNhMS05NGNmLTJkODM0YWFhMWY0YdgCBuACAQ&sid=481ee3b058aa3b0c5fdb0e4ac9521a9c&cc1=sg;dist=1;pagename=the-ritz-carlton-millenia-singapore;srpvid=443b64d3ad1f0192;type=total&&offset={num};rows=25"

    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
    
    # Open the webpage
    driver.get(url)

    # Wait for the dynamic content to load
    time.sleep(3)

    # Get the page source after the content has loaded
    page_source = driver.page_source

    # Close the browser
    driver.quit()

    # Parse the page source with BeautifulSoup
    soup = BeautifulSoup(page_source, 'html.parser')

    # Find all review elements
    reviews = soup.find_all('li', class_='review_list_new_item_block')  # Replace with the actual class

    #Extract review texts
    for review in reviews:
        comment = review.find("span", class_="c-review__body").text
        rating = review.find("div", class_="bui-review-score__badge").text
        #print(f"Rating: {rating}\nComment:\n{comment}\n\n")
        df = df.append({"Hotel": "Ritz Carlton", "Review": comment, "Rating": rating}, ignore_index = True)
        
df.loc[df["Hotel"] == "Ritz Carlton",:]
```

### Categorize the Ratings

```python
df["Rating"] = df["Rating"].astype("float")
df["Rating Category"] = ""

# Filter conditions for setting the rating category
positive = df["Rating"] > 6
neutral = (df["Rating"] >= 5) & (df["Rating"] <= 6)
negative = df["Rating"] < 5

df.loc[positive,"Rating Category"] = "Positive"
df.loc[neutral,"Rating Category"] = "Neutral"
df.loc[negative,"Rating Category"] = "Negative"

df["Rating Category"].value_counts()
```

```python
import matplotlib.pyplot as plt

rating_grouped = df.groupby(by="Rating")["Rating"].count()
rating_grouped.plot.bar(figsize=(16,8))
plt.title("Count By Ratings")
plt.ylabel("Count")
```
